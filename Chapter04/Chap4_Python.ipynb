```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Python Spark dataframe example").config("spark.some.config.option", "some-value").getOrCreate()

sales_df = spark.read.option("sep", "\t").option("header", "true").csv("/FileStore/tables/sample_10000.txt")
# Displays the content of the DataFrame to stdout
sales_df.show()

#Loading data from Parquet
sales_df = spark.read.option("sep", "\t").option("header", "true").csv("/FileStore/tables/sample_10000.txt")

sales_df.write.parquet("sales.parquet")

parquetSales = spark.read.parquet("sales.parquet")

parquetSales.createOrReplaceTempView("parquetSales")
ip = spark.sql("SELECT ip FROM parquetsales WHERE id >= 10 AND id <= 19")
ip.show()

#Dataframe Operations : Printschema
sales_df.printSchema()

#Dataframe Operations : Select
sales_df.select("firstname").show()

#Dataframe Operations : Filter
sales_df.filter(sales_df['id'] < 50).show()

#Dataframe Operations : Group By
sales_df.groupBy("ip").count().show()

#Temporary Views on Dataframes
sales_df.createOrReplaceTempView("sales")
sqlDF = spark.sql("SELECT * FROM sales")
sqlDF.show()

#Global Temporary Views on Dataframes
sales_df.createGlobalTempView("sales")
spark.sql("SELECT * FROM global_temp.sales").show()
spark.newSession().sql("SELECT * FROM global_temp.sales").show()

#Datasets
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Python Spark dataframe example").config("spark.some.config.option", "value").getOrCreate()

#Loading Dataset from CSV with type Sales
ds_sales = spark.read.option("sep", "\t").option("header", "true").csv("/FileStore/tables/sample_10000.txt")
# Displays the content of the Dataset to stdout
ds_sales.show()
```
